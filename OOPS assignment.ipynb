Access: A static method can't access or modify the class state or instance state. It is usually used when you want to perform some operation that is related to the class but doesn't need to access any instance or class-specific data.

Definition: A @classmethod is a method that takes cls as the first parameter, which refers to the class itself, not an instance of the class. This allows the method to access and modify class-level attributes and state.

Access: A class method can modify the class state that applies to all instances of the class. It is often used for factory methods or other methods that need to operate at the class level rather than the instance level.

Use Case: Typically used when the method needs to work with the class itself rather than instances, such as in alternative constructors or methods that affect the class as a whole.

```python
Q3.
# Base class
class Vehicle:
    def __init__(self, vehicle_type):
        self.type = vehicle_type

    def display_info(self):
        print(f"Vehicle Type: {self.type}")

# First derived class
class Car(Vehicle):
    def __init__(self, vehicle_type, brand):
        super().__init__(vehicle_type)
        self.brand = brand

    def display_info(self):
        super().display_info()
        print(f"Car Brand: {self.brand}")

# Second derived class (multi-level)
class ElectricCar(Car):
    def __init__(self, vehicle_type, brand, battery_capacity):
        super().__init__(vehicle_type, brand)
        self.battery = battery_capacity

    def display_info(self):
        super().display_info()
        print(f"Battery Capacity: {self.battery} kWh")

# Example usage
if __name__ == "__main__":
    tesla = ElectricCar("Electric", "Tesla", 100)
    tesla.display_info()
```

```python
Q4.
# Base class
class Bird:
    def fly(self):
        print("Bird is flying...")

# Derived class 1
class Sparrow(Bird):
    def fly(self):
        print("Sparrow flies high in the sky!")

# Derived class 2
class Penguin(Bird):
    def fly(self):
        print("Penguins can't fly, they swim!")

# Polymorphism in action
def show_flight(bird):
    bird.fly()

# Usage
sparrow = Sparrow()
penguin = Penguin()

show_flight(sparrow)
show_flight(penguin)
```

```python
Q5.
class BankAccount:
    def __init__(self, initial_balance=0):
        self.__balance = initial_balance  # Private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: {amount}")
        else:
            print("Invalid deposit amount")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew: {amount}")
        else:
            print("Insufficient balance or invalid amount")

    def check_balance(self):
        print(f"Current Balance: {self.__balance}")

# Example usage
if __name__ == "__main__":
    account = BankAccount(100000)
    account.check_balance()
    account.deposit(50000)
    account.withdraw(30000)
    account.check_balance()
```

```python
Q6.
# Base class
class Instrument:
    def play(self):
        print("Playing an instrument...")

# Derived class: Guitar
class Guitar(Instrument):
    def play(self):
        print("Strumming the guitar!")

# Derived class: Piano
class Piano(Instrument):
    def play(self):
        print("Playing the piano keys!")

# Polymorphism in action
def start_playing(instrument):
    instrument.play()

# Example usage
if __name__ == "__main__":
    instr1 = Guitar()
    instr2 = Piano()

    start_playing(instr1)
    start_playing(instr2)
```

```python
Q7.
class MathOperations:

    # Class method
    @classmethod
    def add_numbers(cls, a, b):
        return a + b

    # Static method
    @staticmethod
    def subtract_numbers(a, b):
        return a - b

# Example usage
if __name__ == "__main__":
    # Using class method
    sum_result = MathOperations.add_numbers(10, 5)
    print(f"Sum: {sum_result}")

    # Using static method
    diff_result = MathOperations.subtract_numbers(10, 5)
    print(f"Difference: {diff_result}")
```

```python
Q10.
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Override the + operator
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"

# Example usage
if __name__ == "__main__":
    v1 = Vector(2, 3)
    v2 = Vector(4, 5)
    v3 = v1 + v2  # Uses overloaded __add__ method

    print(f"v1 = {v1}")
    print(f"v2 = {v2}")
    print(f"v1 + v2 = {v3}")
```

```python
Q12.
class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades

    def average_grade(self):
        if not self.grades:
            return 0
        return sum(self.grades) / len(self.grades)

# Example usage
if __name__ == "__main__":
    student = Student("John", [85, 90, 78, 92])
    print(f"{student.name}'s average grade is {student.average_grade():.2f}")
```

```python
Q14.
class Employee:
    def __init__(self, name, hours_worked, hourly_rate):
        self.name = name
        self.hours_worked = hours_worked

        self.hourly_rate = hourly_rate

    def calculate_salary(self):
        return self.hours_worked * self.hourly_rate

class Manager(Employee):
    def __init__(self, name, hours_worked, hourly_rate, bonus):
        super().__init__(name, hours_worked, hourly_rate)
        self.bonus = bonus

    def calculate_salary(self):
        base_salary = super().calculate_salary()
        return base_salary + self.bonus

# Example usage
if __name__ == "__main__":
    emp = Employee("John", 40, 20)
    mgr = Manager("Alice", 40, 30, 500)

    print(f"{emp.name}'s Salary: {emp.calculate_salary()}")
    print(f"{mgr.name}'s Salary: {mgr.calculate_salary()}")
```

```python
Q16.
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Cow(Animal):
    def sound(self):
        print("Moo")

class Sheep(Animal):
    def sound(self):
        print("Baa")

# Example usage
if __name__ == "__main__":
    cow = Cow()
    sheep = Sheep()

    cow.sound()
    sheep.sound()
```

```python
Q18.
class House:
    def __init__(self, address, price):
        self.address = address
        self.price = price

class Mansion(House):
    def __init__(self, address, price, number_of_rooms):
        super().__init__(address, price)
        self.number_of_rooms = number_of_rooms

# Example usage
if __name__ == "__main__":
    house = House("123 Maple St", 250000)
    mansion = Mansion("456 Oak Ave", 1500000, 10)

    print(f"House at {house.address} costs ${house.price}")
    print(f"Mansion at {mansion.address} costs ${mansion.price} and has {mansion.number_of_rooms} rooms")
```
